```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (adjust path as needed)
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_2590.csv", skiprows=4)

# Extract 2023 population data and remove missing values
population_2023 = df[['Country Name', '2023']].dropna()

# Plot histogram
plt.figure(figsize=(10, 6))
sns.histplot(population_2023['2023'], bins=30, kde=True, color='skyblue')
plt.title("Distribution of Country Populations in 2023")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Top 10 populous countries
top_10 = population_2023.sort_values(by='2023', ascending=False).head(10)

# Bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='2023', y='Country Name', data=top_10, palette='Blues_r')
plt.title("Top 10 Most Populous Countries in 2023")
plt.xlabel("Population")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python
# Bottom 10 least populous countries
bottom_10 = population_2023.sort_values(by='2023', ascending=True).head(10)

# Bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='2023', y='Country Name', data=bottom_10, palette='Greens_r')
plt.title("Bottom 10 Least Populous Countries in 2023")
plt.xlabel("Population")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python
# Box plot
plt.figure(figsize=(10, 3))
sns.boxplot(x=population_2023['2023'], color='orchid')
plt.title("Box Plot: Population Distribution in 2023")
plt.xlabel("Population")
plt.tight_layout()
plt.show()
```

```python
sns.histplot(population_2023['2023'], bins=30, log_scale=True, color='red')
plt.title("Log-Scaled Histogram of Population")
plt.xlabel("Population (Log Scale)")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(population_2023['2023'], bins=30, cumulative=True, kde=True, color='limegreen')
plt.title("Cumulative Histogram of Population")
plt.xlabel("Population")
plt.ylabel("Cumulative Frequency")
plt.show()
```

```python
plt.hist(population_2023['2023'], bins=30, color='magenta', orientation='horizontal')
plt.title("Horizontal Histogram of Population")
plt.ylabel("Population")
plt.xlabel("Frequency")
plt.grid(True)
plt.show()
```

```python
sns.histplot(population_2023['2023'], bins=10, kde=True, color='goldenrod')
plt.title("Histogram with Fewer Bins")
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.show()
```
